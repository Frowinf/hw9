**Credit Default Prediction Modelling**

**Introduction**
Credit default occurs when an individual fails to repay their credit card debts. This issue can cause banks to experience monetary losses. Because of this, being able to successfully predict credit default can be highly beneficial to banks. As data collection improves and the availability of data increases, leveraging machine learning to develop predictive models for credit default becomes more and more beneficial for banks. This project aimed to create a predictive model using machine learning to estimate and predict whether or not a credit card client is likely to default on their credit card debt in the upcoming month.


**The Dataset**
We used the UCI Default of Credit Card Clients dataset. It contains 30,000 records and 24 predictive variables. These include demographic characteristics such as age, gender, marital status, and education, financial indicators such as credit limit, bill statements, and payments over the last six months, and repayment status variables , Pay_0 to Pay_6, which show past delinquencies. Our target variable was default.payment.next.month, a binary outcome which indicates whether or not a customer defaulted in the following billing cycle.
Exploratory data analysis revealed several trends. One thing we learned was that the dataset had a class imbalance, with ~22% of data points being classified as credit default. This can be seen in the visualization below. 
 Exploratory data analysis also revealed that clients with recent payment issues(high pay_0 values), are more likely to default than those without recent payment issues. It also revealed that as people get more educated, their risk of default decreases.

**Modeling**
We began with a DummyClassifier as a baseline model, which simply predicted the majority class (non-default) every time. As expected, it achieved a high accuracy of 0.78,  but an F1 score of 0.00 on the minority class (credit default).
Next, we evaluated the following four machine learning models:
 - Logistic Regression
- K-Nearest Neighbors (KNN)
- Support Vector Classifier (SVC)
- Random Forest (selected as our final model)
 
Each model was embedded in a pipeline for preprocessing. We standardized numerical features and one-hot encoded categorical ones. We also experimented with combining features such as total bill amount and total payments. Hyperparameter tuning was done using RandomizedSearchCV with 5-fold cross-validation while optimizing for F1 score.

**Results**
The Random Forest model outperformed the other models. It achieved a test F1 score of 0.48 and a test accuracy of 0.82. The final confusion matrix showed 745 true positives out of 1,991 defaulters as shown in figure 2. It had  a recall of 0.37 and precision of 0.68 for the default class. These results showed a strong improvement over simpler models, especially the baseline.
To increase interpretability, we analyzed feature importances as seen in figure 3. The top features were:
PAY_0 (most recent payment delay status),
TOTAL_BILL_AMT (total billed amount over six months),
TOTAL_PAY_AMT (cumulative past repayments),
AGE, 
and LIMIT_BAL (credit limit).

**Caveats**
Hard to interpret: While the Random Forest model gave the best results, it's harder to interpret than simpler models. Using a tool such as SHAP could help break down the model and increase interoperability.

Time based data: Our features include several months of payment history. It is possible that the data required for the model to predict credit default is only accessible too late.  There‚Äôs a risk of using information that might not be available until the credit default is already bound to happen.

Fairness concerns: Features like education,marital status, or age could unintentionally reflect social biases. If the model were to be used in real life application the model should be carefully reviewed to make sure there are no unfair biases.
